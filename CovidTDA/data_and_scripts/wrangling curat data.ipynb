```python
## Make BMI into categories
def bmi(row):
    if row["BMI"] < 18.5:
        return "underweight"
    if row["BMI"] < 25:
        return "normal"
    if row["BMI"] < 30:
        return "overweight"
    if row["BMI"] >= 30:
        return "obese"
    
metadata["BMI"] = metadata.apply(lambda row: bmi(row), axis=1)
metadata.BMI.value_counts()
```

```python
## excluding all children and NAs
print(metadata2.age.min())

A1 = metadata2.age_category != "newborn"
A2 = metadata2.age_category.isna()
metadata2 = metadata2[A1 | A2] # | is a bitwise OR operator
metadata2 = metadata2[~metadata2.age.isin(range(2))]  # this doubling is necessary as some don't have category but are under 18.
#now make sure noone has no age value
metadata2 = metadata2[~((metadata2['age'].isnull()) & (metadata2['age_category'].isnull()))]

print(metadata2.age.min())
```

```python
### calculating weighted mean
final_samples = pd.Series(metadata2.country.value_counts())
final_samples
```

```python
final_samples_df = pd.DataFrame({'country':final_samples.index, 'final_samples':final_samples.values})
covid = pd.merge(covid, final_samples_df, on = "country")
```

```python
# actual weighted mean
def w_avg(df, values, weights):
    d = df[values]
    w = df[weights]
    return (d * w).sum() / w.sum()
threshold = w_avg(covid, 'hosp_by_pop', 'final_samples')
threshold
```

```python
print("LOW")
print(list(covid[covid.hosp_by_pop < threshold].index))
print("HIGH")
print(list(covid[covid.hosp_by_pop > threshold].index))
```

```python
#make low/high groups
# threshold = 1848.5856162414661
metadata2['group'] = metadata2.hosp_by_pop.map(lambda x: 'high' if x>threshold else 'low')
metadata2
```

```python
## Dummy code, sex, BMI, and low/high
cat_columns = ["gender", "BMI", "group", "age_category"]
metadata_dummies = pd.get_dummies(metadata2, prefix_sep=": ", columns=cat_columns)
```

```python
# save file
metadata_dummies.to_csv("curat_metadata_filtered.csv")
```

Filtering the OTU table to only include subjects in metadata

```python
otu = pd.read_csv("curat_otu_table.csv", index_col=0, encoding='utf-16-le')
otu = otu.transpose()
otu = otu.reset_index()
otu = otu.rename(columns={"index": "ppt_id"})
#otu = otu.set_index("ppt_id")
otu
```

```python
## make otu and metadata include the same samples
# need to open metadata again so that it doesn't have subjectID as index but as column
metadata_final = pd.read_csv("curat_metadata_filtered.csv")

otu = otu[otu["ppt_id"].isin(metadata_final["ppt_id"].to_list())]
otu = otu.set_index("ppt_id")
print(len(metadata_final) == len(otu))
otu
```

```python
### Remove columns of taxa which are empty

otu.loc['sum'] = otu.sum(axis=0)
otu
```

```python
# for those that have sum == 0, drop the column, first find column names
empty_taxa = []

for column in otu.columns:
    column_series = otu[column]
    if column_series.iloc[-1] == 0:
        empty_taxa.append(column)
print(len(empty_taxa))
        
# drop columns, drop sum row
otu = otu.drop(columns = empty_taxa)
otu.drop(otu.tail(1).index,inplace=True)
```

```python
otu
```

```python
# get rid of leading characters in the columns
newNames = []
for col in otu.columns:
    newNames.append(col.lstrip("s__"))
otu.columns = newNames
```

```python
# overwrite files
otu.to_csv("curat_otu_filtered.csv")
```
